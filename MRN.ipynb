```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
amin = 50
amax = 2500
```

```python
asize = np.logspace(np.log10(amin), np.log10(amax))
```

```python
C_carb = 10.**(-25.13)
C_sil  = 10.**(-25.11)

nH = 1.0
```

```python
dn_carb = C_carb*asize**(-3.5)
dn_sil  = C_sil *asize**(-3.5)
```

```python
fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(111)

ax.plot(asize, dn_carb*1.0e29, "-r", linewidth=2)
ax.plot(asize, dn_sil *1.0e29, "-k", linewidth=2)

ax.set_xscale("log")
ax.set_yscale("log")

#ax.set_xlabel("grain size [$\\AA$]", fontsize=20)
#ax.set_ylabel("10$^{29}$ n$_{\mathrm{H}}^{-1}$ dn$_{\mathrm{gr}}$/da", fontsize=20)

ax.tick_params(axis='both', which='major', length=10, width=2,  labelsize=15, direction="in")
ax.tick_params(axis='both', which='minor', length=5, width=1.5, labelsize=15, direction="in")

fig.show()
```

```python
fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(111)

ax.plot(asize, dn_carb*(asize)**4*1.0e-4, "-r", linewidth=2)
ax.plot(asize, dn_sil *(asize)**4*1.0e-4, "-k", linewidth=2)

ax.set_xscale("log")
ax.set_yscale("log")

ax.set_xlabel("grain size [$\\AA$]", fontsize=20)
#ax.set_ylabel("10$^{29}$ n$_{\\mathrm{H}}^{-1}$ a$^{4}$ dn$_{\\mathrm{gr}}$/da", fontsize=20)

ax.tick_params(axis='both', which='major', length=10, width=2,  labelsize=15, direction="in")
ax.tick_params(axis='both', which='minor', length=5, width=1.5, labelsize=15, direction="in")

#ax.set_ylim(0.1, 1.0e2)

fig.show()
```
